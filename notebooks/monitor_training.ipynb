# Plot the training loss and gradients

```python
import json
import os

# unroll the metrics
import matplotlib.pyplot as plt
import pandas as pd

from utils import load_json

run_path = "my_runs/8"


metrics = load_json(os.path.join(run_path, "metrics.json"))
configs = load_json(os.path.join(run_path, "config.json"))
print(f"Configs: {configs}")


fig, axs = plt.subplots(2, 2, figsize=(10, 10))
for i, (k, v) in enumerate(metrics.items()):
    axs[i // 2, i % 2].plot(v["values"])
    axs[i // 2, i % 2].set_title(k)
```
